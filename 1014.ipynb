```python
import numpy as np
from sklearn import  datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
data
```

```python
target
```

```python
np.random.seed(29)

raw_df
X = data
Y = target

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
regression_clf = LinearRegression()
regression_clf.fit(X_train,Y_train)
Y_predict = regression_clf.predict(X_test)
scort = r2_score(Y_test,Y_predict)
print("房價的預測準確率:",scort)

plt.plot(Y_test,label = 'Real Price')
plt.plot(Y_predict,label = 'Predict Price')
plt.legend()
```

```python
MSE = np.square(np.subtract(Y_test,Y_predict)).mean()
MSE
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,Y_predict)
```
